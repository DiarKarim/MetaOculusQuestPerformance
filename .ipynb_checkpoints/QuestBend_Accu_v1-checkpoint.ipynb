```python
def unit_vector(vector):
    """ Returns the unit vector of the vector.  """
    return vector / np.linalg.norm(vector)

def angle_between(v1, v2):
    """ Returns the angle in radians between vectors 'v1' and 'v2'::

            >>> angle_between((1, 0, 0), (0, 1, 0))
            1.5707963267948966
            >>> angle_between((1, 0, 0), (1, 0, 0))
            0.0
            >>> angle_between((1, 0, 0), (-1, 0, 0))
            3.141592653589793
    """
    v1_u = unit_vector(v1)
    v2_u = unit_vector(v2)
    
    rad = np.arccos(np.clip(np.dot(v1_u, v2_u), -1.0, 1.0))
    deg = math.degrees(rad)
    
    return deg
```

```python
#--------------- Cross-Correlation ---------------------------------------------

def CrossCorr(vel_1, vel_2, sampleFreq):    
    
    corr = np.correlate(vel_1 - np.mean(vel_1), 
                      vel_2 - np.mean(vel_2),
                      mode='full')
    
    sampleDifference = np.argmax(vel_2) - np.argmax(vel_1) # What is this 20? np.argmax(vel_2[20:]) - np.argmax(vel_1[20:])

    lag = (sampleDifference  * sampleFreq) * 1000
    
    return lag
```

```python
def ResizeArray(data, newSize):
    x = data
    i = newSize
    z = i / len(x)
    x_int = interpolation.zoom(x,z)

    return x_int

def AngularCorrection(data):
    outputarr = []
    for i in data:
        if i>180:
            outputarr.append(i - 270)
        else:
            outputarr.append(i + 90)

    return outputarr

def AverageCurve(data,col):
    # plt.figure()
    ydata = np.mean(data,axis=0)
    xvls = np.linspace(0,len(ydata),len(ydata)) 
    yerr = np.std(data, axis=0) / np.sqrt(np.shape(data)[0])
    plt.plot(xvls,ydata,color=col)
    plt.fill_between(xvls, ydata-yerr, ydata+yerr, alpha=0.5,color=[0.1,0.1,0.1])

# marker='s', mfc='red', mec='red', ms=5, mew=2,

def PlotErrorBars(dataX = np.tile(np.nan,10), dataY = np.tile(np.nan,10), dataZ = np.tile(np.nan,10), colorz = 'r'):
    # Clean up outliers 5x outside the mean  
    xData =  [(i * np.nan) if i > (np.nanmean(dataX) * 5.0) else i for i in dataX]
    yData =  [(i * np.nan) if i > (np.nanmean(dataY) * 5.0) else i for i in dataY]
    zData =  [(i * np.nan) if i > (np.nanmean(dataZ) * 5.0) else i for i in dataZ]

    # Compute standard errors
    x_SE  = np.std(xData, axis=0) / np.sqrt(np.shape(xData)[0])
    y_SE  = np.std(yData, axis=0) / np.sqrt(np.shape(yData)[0])
    z_SE  = np.std(zData, axis=0) / np.sqrt(np.shape(zData)[0])
    all_SE = [x_SE, y_SE, z_SE]

    # Plot data 
    plt.errorbar([0,1,2], [np.nanmean(xData),np.nanmean(yData),np.nanmean(zData)], all_SE, color = colorz, marker='s')


def PlotErrorBars2(dataX = np.tile(np.nan,10), colorz = 'r'):
    # Clean up outliers 5x outside the mean  
    xData =  [(i * np.nan) if i > (np.nanmean(dataX) * 5.0) else i for i in dataX]

    # Compute standard errors
    x_SE  = np.std(xData, axis=0) / np.sqrt(np.shape(xData)[0])

    # Plot data 
    plt.errorbar([0], np.nanmean(xData), x_SE, color = colorz, marker='s')
```

```python
def SamplingRate(timeValues):
    # Compute actual sampling rate
    timetaken = timeValues
#     timetaken2 = ResizeArray(timetaken, newArrLength)
    timetaken3 = np.round(timetaken,1)
    timetaken3 = timetaken3.tolist()

    # If 0.0 time isn't present, then use the smallest time value as the start of the trial
    try:
        indexOfStart = timetaken3.index(0.0) # indexOfStart = np.where(timetaken == 0.0)
    except Exception as e:
        print('IndexErr: ', e)
        minTimeVal = np.nanmin(timetaken3)
        indexOfStart = timetaken3.index(minTimeVal)

    # Check if devision by zero, because the last value happens to be zero and use last largest value instead
    try:
        samplingRate = np.round(1.0 / ((len(timetaken3)-indexOfStart) / timetaken3[-1]), 4)
    except Exception as e:
        print('SampleErr: ', e)
        lastMaxTimeVal = np.nanmax(timetaken3[-1-10:-1])
        samplingRate = np.round(1.0 / ((len(timetaken3)-indexOfStart) / lastMaxTimeVal), 4)

#     print('Sampling Rate: ', np.round(1.0/samplingRate), ' Hz')
    
    return np.round(1.0/samplingRate)
```

```python
user_tempo_80 = '1634123474'
user_tempo_120 = '1634123557'
user_tempo_160 = '1634123606'

tempoUsers = [user_tempo_80, user_tempo_120, user_tempo_160]
tempos = ['80','120','160']

df_results = None
df_raw_angs = None
df_rawVel = None

a_1_2,a_2_3,a_3_4 = [],[],[]
v_1_2,v_2_3,v_3_4 = [],[],[]

for idx, tempoUser in enumerate(tempoUsers): 

    realMask1 = (df_allBend['UserID'] == tempoUser) & (df_allBend['gameObjectName'] == 'realFingerTip')
    realMask2 = (df_allBend['UserID'] == tempoUser) & (df_allBend['gameObjectName'] == 'realFingerTip (1)')
    realMask3 = (df_allBend['UserID'] == tempoUser) & (df_allBend['gameObjectName'] == 'realFingerTip (2)')
    realMask4 = (df_allBend['UserID'] == tempoUser) & (df_allBend['gameObjectName'] == 'realFingerTip (3)')

    virtMask1 = (df_allBend['UserID'] == tempoUser) & (df_allBend['gameObjectName'] == 'r_index_fingernail_marker')
    virtMask2 = (df_allBend['UserID'] == tempoUser) & (df_allBend['gameObjectName'] == 'b_r_index1')
    virtMask3 = (df_allBend['UserID'] == tempoUser) & (df_allBend['gameObjectName'] == 'b_r_index2')
    virtMask4 = (df_allBend['UserID'] == tempoUser) & (df_allBend['gameObjectName'] == 'b_r_index3')

    # Real 
    # [angleCorrected_1_2, angleCorrected_2_3, angleCorrected_3_4]
    df_real = ComputeAngles(realMask1,realMask2,realMask3,realMask4,'k')
    df_real.insert(0,'Realm', 'Real', True)
#     print('Real')
#     print('Mean bend ang: ', np.nanmean(df_real['Angle_m1_m2']))
#     print('Min bend ang: ', np.min(df_real['Angle_m1_m2']), '\nMax bend ang: ', np.max(df_real['Angle_m1_m2']))
#     print('Range ang: ', np.diff([np.min(df_real['Angle_m1_m2']), np.max(df_real['Angle_m1_m2'])]), '\n')

    # Virtual 
    # [VangleCorrected_1_2, VangleCorrected_2_3, VangleCorrected_3_4] 
    df_virt = ComputeAngles(virtMask1,virtMask4,virtMask3,virtMask2,'m')
    df_virt.insert(0,'Realm', 'Virtual', True)
#     print('Virtual')
#     print('Mean bend ang: ', np.nanmean(df_virt['Angle_m1_m2']))
#     print('Min bend ang: ', np.min(df_virt['Angle_m1_m2']), '\nMax bend ang: ', np.max(df_virt['Angle_m1_m2']))
#     print('Range ang: ', np.diff([np.min(df_virt['Angle_m1_m2']), np.max(df_virt['Angle_m1_m2'])]))

    df_all = pd.concat((df_real, df_virt))

    sI = 440
    eI = 500

    # Real 
    plt.figure()
    ax1 = plt.axes(projection='3d')
    plt.title('Real')
    mask_real = (df_all['Realm'] == 'Real')
    ax1.plot3D(df_all[mask_real].Marker_1_posX.values[sI:eI], df_all[mask_real].Marker_1_posY.values[sI:eI], df_all[mask_real].Marker_1_posZ.values[sI:eI],'k-o')
    ax1.plot3D(df_all[mask_real].Marker_2_posX.values[sI:eI], df_all[mask_real].Marker_2_posY.values[sI:eI], df_all[mask_real].Marker_2_posZ.values[sI:eI],'k-o')
    ax1.plot3D(df_all[mask_real].Marker_3_posX.values[sI:eI], df_all[mask_real].Marker_3_posY.values[sI:eI], df_all[mask_real].Marker_3_posZ.values[sI:eI],'k-o')
    ax1.plot3D(df_all[mask_real].Marker_4_posX.values[sI:eI], df_all[mask_real].Marker_4_posY.values[sI:eI], df_all[mask_real].Marker_4_posZ.values[sI:eI], 'k-o')

    # Virtual 
    # plt.figure()
    # ax2 = plt.axes(projection='3d')
    plt.title('Virtual')
    mask_virt = (df_all['Realm'] == 'Virtual')
    ax1.plot3D(df_all[mask_virt].Marker_1_posX.values[sI:eI], df_all[mask_virt].Marker_1_posY.values[sI:eI], df_all[mask_virt].Marker_1_posZ.values[sI:eI],'m-o')
    ax1.plot3D(df_all[mask_virt].Marker_2_posX.values[sI:eI], df_all[mask_virt].Marker_2_posY.values[sI:eI], df_all[mask_virt].Marker_2_posZ.values[sI:eI],'m-o')
    ax1.plot3D(df_all[mask_virt].Marker_3_posX.values[sI:eI], df_all[mask_virt].Marker_3_posY.values[sI:eI], df_all[mask_virt].Marker_3_posZ.values[sI:eI],'m-o')
    ax1.plot3D(df_all[mask_virt].Marker_4_posX.values[sI:eI], df_all[mask_virt].Marker_4_posY.values[sI:eI], df_all[mask_virt].Marker_4_posZ.values[sI:eI], 'm-o')

    plt.rcParams["figure.figsize"] = (12,4)

    plt.figure()

    plt.subplot(131)
    plt.title('Angle Between Joints DP-IP')
    plt.plot(df_all[mask_real].Angle_m1_m2,'k')
    plt.plot(df_all[mask_virt].Angle_m1_m2,'m')
    plt.legend(['Real', 'Virtual'])

    plt.subplot(132)
    plt.title('Angle Between Joints IP-PP')
    plt.plot(df_all[mask_real].Angle_m2_m3,'k')
    plt.plot(df_all[mask_virt].Angle_m2_m3,'m')

    plt.subplot(133)
    plt.title('Angle Between Joints PP-MC')
    plt.plot(df_all[mask_real].Angle_m3_m4,'k')
    plt.plot(df_all[mask_virt].Angle_m3_m4,'m')

    times_real = df_all[mask_virt].Time

    plt.figure()
    plt.rcParams["figure.figsize"] = (12,5)

    plt.subplot(311)
    print('Distal-Medial')
    ang_real_1_2 = df_all[mask_real].Angle_m1_m2
    ang_virt_1_2 = df_all[mask_virt].Angle_m1_m2
    df_1_2, ang_1_2, angVel_1_2, angErr_Arr_NoLag_1_2, angVelArr_NL_1_2 = PlotMetrics(ang_real_1_2, ang_virt_1_2, times_real, 'Distal-Medial')
    a_1_2.append(ang_1_2)
    v_1_2.append(angVel_1_2)

    plt.subplot(312)
    print('Medial-Proximal')
    ang_real_2_3 = df_all[mask_real].Angle_m2_m3
    ang_virt_2_3 = df_all[mask_virt].Angle_m2_m3
    df_2_3, ang_2_3, angVel_2_3, angErr_Arr_NoLag_2_3, angVelArr_NL_2_3 = PlotMetrics(ang_real_2_3, ang_virt_2_3, times_real, 'Medial-Proximal')
    a_2_3.append(ang_2_3)
    v_2_3.append(angVel_2_3)
    
    plt.subplot(313)
    print('Proximal-Metacapal')
    ang_real_3_4 = df_all[mask_real].Angle_m3_m4
    ang_virt_3_4 = df_all[mask_virt].Angle_m3_m4
    df_3_4, ang_3_4, angVel_3_4, angErr_Arr_NoLag_3_4, angVelArr_NL_3_4 = PlotMetrics(ang_real_3_4, ang_virt_3_4, times_real, 'Proximal-Metacapal')
    a_3_4.append(ang_3_4)
    v_3_4.append(angVel_3_4)
    
    tmpDF = pd.concat(( [df_1_2, df_2_3, df_3_4] ))
    tmpDF.insert(0,'Tempo', tempos[idx])
    
    if df_results is None:
        df_results = tmpDF
    else:
        df_results = pd.concat((df_results, tmpDF))
    
    # ------------------------------ Ang Error Array -----------------------------------------
    tmpDFang = pd.concat((angErr_Arr_NoLag_1_2, angErr_Arr_NoLag_2_3, angErr_Arr_NoLag_3_4))
    tmpDFang.insert(0,'Tempo', tempos[idx])
    
    if df_raw_angs is None:
        df_raw_angs = tmpDFang
    else:
        df_raw_angs = pd.concat((df_raw_angs, tmpDFang))
        
    # ------------------------------ Ang Vel Error Array -----------------------------------------
    tmpDFang2 = pd.concat((angVelArr_NL_1_2, angVelArr_NL_2_3, angVelArr_NL_3_4))
    tmpDFang2.insert(0,'Tempo', tempos[idx])
    
    if df_rawVel is None:
        df_rawVel = tmpDFang2
    else:
        df_rawVel = pd.concat((df_rawVel, tmpDFang2))     
```
